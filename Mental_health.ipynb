```python
label = LabelEncoder()
df['gender'] = label.fit_transform(df['gender'])
df['employment_status'] = label.fit_transform(df['employment_status'])
df['work_environment'] = label.fit_transform(df['work_environment'])
df['work_environment'] = label.fit_transform(df['work_environment'])
df['mental_health_history'] = label.fit_transform(df['mental_health_history'])
df['seeks_treatment'] = label.fit_transform(df['seeks_treatment'])
df['mental_health_risk'] = label.fit_transform(df['mental_health_risk'])
```

```python
x, y = df.iloc[:, 0:-1], df.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3, shuffle=True, random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
df.head()
```

```python
x_train_tensor = torch.tensor(x_train, dtype=torch.float32)
x_test_tensor = torch.tensor(x_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.to_numpy(), dtype=torch.long)
y_test_tensor = torch.tensor(y_test.to_numpy(), dtype=torch.long)
```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self, input_size, output):
        super(NeuralNetwork, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_size, 128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU(),
            nn.Dropout(p=0.2),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, num_classes)
        )
    def forward(self, x):
        return self.network(x)
            
```

```python
input_size = x_train.shape[1]
hidden_size = 64
num_classes = len(np.unique(y))
model = NeuralNetwork(input_size, num_classes)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 200
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")

    model.train()
    outputs = model(x_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    _, predicted = torch.max(outputs, 1)
    correct = (predicted == y_train_tensor).sum().item()
    accuracy = correct / y_train_tensor.size(0)

    print(f"Loss: {loss.item():.4f}, Accuracy: {accuracy * 100:.2f}%\n")

print("Done!")
```

```python
model.eval()
```

```python
# Set the model to evaluation mode
model.eval()

# Turn off gradients during evaluation for efficiency
with torch.no_grad():
    # Get predictions for test data
    test_outputs = model(x_test_tensor)

    # Get predicted class labels
    _, test_predicted = torch.max(test_outputs, 1)

    # Calculate accuracy
    test_correct = (test_predicted == y_test_tensor).sum().item()
    test_accuracy = test_correct / y_test_tensor.size(0)

    print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
